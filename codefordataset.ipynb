```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Loading the dataset

data = pd.read_csv('president_heights.csv')
data.head
```

```python
# Preparing the data for linear regression
X = data['order'].values.reshape(-1, 1)  # Independent variable (Presidential Order)
Y = data['height(cm)'].values  # Dependent variable (Height)

# Creating the linear regression model
model = LinearRegression()
model.fit(X, Y)

# Making predictions
Y_pred = model.predict(X)

# Calculating metrics
mse = mean_squared_error(Y, Y_pred)
r2 = r2_score(Y, Y_pred)

# Plotting the results
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Fitted Line')
plt.title('Linear Regression of Presidential Order vs Height')
plt.xlabel('Presidential Order')
plt.ylabel('Height (cm)')
plt.legend()
plt.show()

# Outputting the model coefficients
print("Intercept (β0):", model.intercept_)
print("Slope (β1):", model.coef_[0])
print("Mean Squared Error (MSE):", mse)
print("R-squared (R²):", r2)
```
